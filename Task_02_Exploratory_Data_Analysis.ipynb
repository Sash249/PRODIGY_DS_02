## Objective
To perform exploratory data analysis on the Titanic dataset to understand
patterns, relationships, and factors affecting survival.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.info()
df.describe()
```

```python
# Fill missing age values with median
df['age'].fillna(df['age'].median(), inplace=True)

# Drop deck column (too many missing values)
df.drop(columns=['deck'], inplace=True)

df.isnull().sum()

```

```python
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()

```

```python
sns.countplot(x='survived', hue='sex', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
plt.hist(df['age'], bins=20)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

## Observations
- Females had a higher survival rate compared to males.
- First-class passengers had better survival chances.
- Younger passengers showed slightly higher survival rates.
- Most passengers were aged between 20 and 40.

## Conclusion
The EDA shows that gender, passenger class, and age significantly influenced
survival outcomes in the Titanic dataset.
